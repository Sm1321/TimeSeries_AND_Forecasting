### **Handling Missing Values**

### **fillna() -method**

- Front Filling
 - Assigens the value of the Previous Period


 - Back Filling
  - Assigns the Value for the next Period


  - Assigning the Value same Value
   - Assign the Average to all the Missing Values within the time-series.

```python
df_cpy['spx'] = df_cpy['spx'].fillna(method='ffill')  # Forward fill missing values
```

```python
df_cpy.isna().sum()
```

```python
df_cpy['ftse'] = df_cpy['ftse'].fillna(method='bfill')  # Backward fill missing values
```

```python
df_cpy.isna().sum()
```

```python
df_cpy['dax'] = df_cpy['dax'].fillna(value = df_cpy['dax'].mean())
```

```python
df_cpy.isna().sum()
```
